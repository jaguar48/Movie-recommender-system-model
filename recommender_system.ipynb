```python
credits = credits.rename(index=str,columns= {'movie_id':'id'})
merged = movies.merge(credits, on='id')
```

```python
merged.head()
```

```python
merged.drop(['homepage','status','title_x','title_y'],axis=1,inplace=True)
```

```python
merged.info()
```

```python
R= merged['vote_average']
v= merged['vote_count']
C= merged['vote_average'].mean()
m= merged['vote_count'].quantile(.70)
```

```python
merged['Weighted_avg']= ((R*v)+(C*m))/(v+m)
```

```python
merged.head()
```

```python
movies_ranking = merged.sort_values('Weighted_avg',ascending=False)
movies_ranking.head()
```

```python
movies_ranking[['original_title','vote_count','vote_average','Weighted_avg','popularity']].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
merged.isnull().sum()
```

```python
merged['overview']=merged['overview'].fillna('')
merged['overview']
```

```python
tfidf=TfidfVectorizer(strip_accents='ascii',stop_words='english',analyzer='word',token_pattern=r'\w{1,}',
                      ngram_range=(1,3),min_df=3)
```

```python
tfidf_matrix=tfidf.fit_transform(merged['overview'])
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
sig = sigmoid_kernel(tfidf_matrix,tfidf_matrix)
```

```python
indices=pd.Series(merged.index,index=merged['original_title']).drop_duplicates()
indices.head(5)
```

```python
def give_rec(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies 
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    movie_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return merged['original_title'].iloc[movie_indices]
```

```python
give_rec('Avatar')
```
